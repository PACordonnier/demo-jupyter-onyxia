```python
import pandas as pd
import plotly.express as px

# Utilisation d'un dataset intégré à Plotly pour éviter de gérer des fichiers CSV
df = px.data.gapminder()

# Une manipulation simple de données
france_data = df[df['country'] == 'France']

# Un graphique interactif
fig = px.line(france_data, x='year', y='lifeExp', 
              title='Évolution de l\'espérance de vie en France')
fig.show()
```
